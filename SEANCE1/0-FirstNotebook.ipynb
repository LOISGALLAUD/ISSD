## Exercice 4
### Jeu simple de données


```python
# Définition d'un jeu simple de données: les températures moyennes par mois à Strasbourg (src: climatdestrasbourg.fr)
temp = [1.9 , 2.9, 7, 10.5, 15, 18.1, 20.1, 19.8, 15.8, 11.2,
5.8, 2.8]
mois = ['Janvier', 'Fevrier', 'Mars', 'Avril', 'Mai', 'Juin',
'Juillet', 'Aout', 'Septembre', 'Octobre', 'Novembre', 'Decembre']

# On précise l'index de la Serie pandas
s = pd.Series(temp, index=mois)
```

```python
# On remarque que lors de l'affichage les données sont alignées avec 
# l'index précisé en argument de la méthode Series
print(s)
```

```python
s.index
```

```python
# Que se passe-t-il si on ne précise pas l'index?
s2 = pd.Series(temp)
```

```python
# On remarque que par défault l'index affiché est [0, ..., n]
print(s2)
```

```python
# Pour accéder aux valeurs des index on utilise:
print(s['Janvier'])
print(s2[0]) # /!\ L'index est un entier
```

## Exercice 5
### Manipulation de données avec les *DataFrame* de pandas

```python
# On définit les températures moyennes de chaque de l'année 2018:
temp2018 = [6.7, 1.0, 5.5, 14.6, 17.7, 20.3, 22.7, 22.4, 17.7, 12.5, 6.7, 5.2]
```

```python
# On crée une DataFrame avec la syntaxe suivante:
df = pd.DataFrame({'Moyenne': temp, 'temp2018': temp2018}, index=mois)
print(df)
print(df.index)
print(df.columns)
```

```python
# Il est possible d'ajouter une nouvelle colonne à la DataFrame:
df['Difference'] = ['+', '-','-', '+', '+','+','+','+','+','+','+','+']
print(df)
print(df.dtypes)
```

```python
# On peut récupérer une colonne de la DataFrame comme ceci:
s3 = df['temp2018']
# ou de manière équivalente: s3 = df.temp2018
print(s3)
print(type(s3))
```

```python
# Pour récupérer plusieurs colonnes en même temps:
df2 = pd.DataFrame(df, columns=['temp2018', 'Difference'])
print(df2)
print(df2.columns)
```

```python
# Pour récupérer une portion de données:
df3 = df['Fevrier':'Mai'] # ou de manière équivalente df3 = df[1:5]
print(df3)
```
